```python
plt.figure(figsize=(16, 6))
features = data.loc[:,"alcohol_reference":"violence"]
fig = sns.heatmap(features.corr(),annot=True,cmap='RdYlGn',linewidths=0.2)
```

```python
data["esrb_rating"].unique()
```

```python
from sklearn import preprocessing 
label_encoder = preprocessing.LabelEncoder()
data['esrb_rating']= label_encoder.fit_transform(data['esrb_rating']) 
  
data['esrb_rating'].unique()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = features.values
y = data["esrb_rating"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
#model.add(Dense(128, activation='relu'))
#model.add(Dense(64, activation='relu'))
model.add(Dense(16, activation='relu'))
#model.add(Dense(16, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', 
              optimizer='sgd',
              metrics=['accuracy'])
```

```python
results = model.fit(x=X_train, y=y_train, epochs=250)
results
```

```python
print("Evaluating on training set...")
(loss, accuracy) = model.evaluate(X_train, y_train, verbose=0)
print("loss={:.4f}, accuracy: {:.4f}%".format(loss,accuracy * 100))

print("Evaluating on testing set...")
(loss, accuracy) = model.evaluate(X_test, y_test, verbose=0)
print("loss={:.4f}, accuracy: {:.4f}%".format(loss,accuracy * 100))
```

```python
# summarize history for accuracy
plt.plot(results.history['accuracy'])
plt.plot(results.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'])
plt.show()

# summarize history for loss
plt.plot(results.history['loss'])
plt.plot(results.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'])
plt.show()


max_loss = np.max(results.history['loss'])
min_loss = np.min(results.history['loss'])
print("Maximum Loss : {:.4f}".format(max_loss))
print("")
print("Minimum Loss : {:.4f}".format(min_loss))
print("")
print("Loss difference : {:.4f}".format((max_loss - min_loss)))
```
